```python
# program to take two numbers as an input and performing all operations 
first_number= int(input("Chal pehla number enter kar"))
second_number= int(input("Chal second number enter kar"))
operation = input("Chal operation enter kar")
if operation == "+":
    result = first_number + second_number
    print(result)
elif operation == "-":
    result = first_number - second_number
    print(result)
elif operation == "*":
    result = first_number * second_number
    print(result)
elif operation == "/":
    result = first_number / second_number
    print(result)
elif operation == "//":
    result = first_number // second_number
    print(result)
elif operation == "%":
    result = first_number % second_number
    print(result)
else:
    print("Sahi operation enter kar gadhe")
```

```python
#wap to use random library to guess the jackpot number and generate it using random
#alsocount the number of guess
import random
jackpot = random.randint(1,100)
guess = int(input("Chal number enter kar"))
count = 1
while guess!= jackpot:
    if guess<jackpot:
        print("thoda bada number guess kar")
        guess = int(input("Chal wapas guess kar"))
        count+=1
    else:
        print("Thoda chota number guess kar")
        guess = int(input("Chal wapas guess kar"))
        count+=1
else:
    print("Correct guess! 10 crore!!!")
    print("Attempts taken to correct guess=",count)
```

# Literals
whatever we store inside a variable is known as literal.

```python
a = 0b100 #Binary literal
b = 100  #Decimal literal
c = 0o310 # Octal literal
d = 0x12C # Hexa decimal literal
print(a,b,c,d)
```

```python
#Float literal
float_1=10.5
float_2=1.5e2 # 1.5*10^2
float_3=1.5e-3 #1.5*10^-3
print(float_1,float_2,float_3)
```

```python
#complex Literal 
a = 3.14j
print(a,a.imag,a.real)
```
