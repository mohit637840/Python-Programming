```python
# ATM machine --> pin,balance
class Atm:
    #Constructor - speacial funct inside class constructor initialize on object creation..
    def __init__(self):
        #jo bhi variables banayenge inside class uhh have to craete them inside the constructor
        self.pin='' # jo bhi variable banane hain uske aage self. lagega
        self.balance=0
        self.menu()
        # print("Main to execute ho gya")
    #object can acess the class so we need to create an object
    # functionality : pinchange,withdraw,bankbalancedisplay,pinset,menu\
    def menu(self):
        user_input= int(input("""Hi How can I help you?
              1. Press 1 to create pin
              2. press 2 to change pin
              3. press 3 to check balance
              4. press 4 to withdraw
              5. anything else to exit 
              """))
        if user_input==1:
            #create pin
            self.create_pin()
        elif user_input==2:
            #change pin
            self.change_pin()
        elif user_input==3:
            #check balance
            self.check_balance()
        elif user_input==4:
            self.withdraw()
            pass 
        else:
            exit()
    def create_pin(self):
        user_pin=input("enter ur pin")
        self.pin=user_pin

        user_balance=int(input("enter balance"))
        self.balance=user_balance
        print("pin created sucessfully")
        self.menu()
    def change_pin(self):
        old_pin=input("Enter old pin")
        if old_pin==self.pin:
            new_pin=input("Enter new pin")
            self.pin=new_pin
            print("pin change sucessfully")
            self.menu()
        else:
            print("Nahi karne de sakta re baba!")
            self.menu()
    def check_balance(self):
        user_pin=input("Enter ur pin")
        if user_pin==self.pin:
            print("Balance:",self.balance)
            self.menu()
        else:
            print("Chal nikal yahan se")
            self.menu()
    def withdraw(self):
        user_pin=input("Enter ur pin")
        if user_pin==self.pin:
            withdraw_amount=int(input("enter amount to withdraw"))
            if withdraw_amount<self.balance:
                self.balance-=withdraw_amount
                print("withdrawn",withdraw_amount,"Sucessfully","account balance =",self.balance)
                self.menu()
            else:
                print("Chal re nikal rasta naap!")
                self.menu()
        else:
            print("Chal nikal yahan se")
            self.menu()
```

```python
class Fraction:
    #parametrized constructor --> expecting some parameter
    def __init__(self,x,y):
        self.num=x
        self.den=y
# at this point python does not know how its object looks like so we use second magic method
    def __str__(self): # whenever we are trying to print something then str ka code execute ho jayega
        return '{}/{}'.format(self.num,self.den) # string formatting 
    def __add__(self,other): #two fraction objects
        new_num=self.num*other.den + other.num*self.den 
        new_den=self.den*other.den
        return '{}/{}'.format(new_num,new_den)
    def __sub__(self,other): #two fraction objects
        new_num=self.num*other.den - other.num*self.den  
        new_den=self.den*other.den
        return '{}/{}'.format(new_num,new_den)
    def __mul__(self,other): #two fraction objects
        new_num=self.num*other.num  
        new_den=self.den*other.den
        return '{}/{}'.format(new_num,new_den)
    def __truediv__(self,other): #two fraction objects can be fr1 and fr2 too
        new_num=self.num*other.den  
        new_den=self.den*other.num
        return '{}/{}'.format(new_num,new_den)
    def covert_to_decimal(self): #non magic method we created it 
        return self.num/self.den
fr1=Fraction(5,7)
print(fr1)
fr2=Fraction(7,8)
print(fr1+fr2)
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
fr1.covert_to_decimal()
```
