### Exceptions

If things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened.

- Exceptions are raised by python runtime
- You have to takle is on the fly
- compiler can not detect that

#### **Examples**

- Memory overflow (for eg. in deep learning we are woking on large data and while loading that it could throw memory memory overflow error)
- Divide by 0 -> logical error
- Database error

```python
# Why is it important to handle exceptions
# how to handle exceptions
# -> Try except block
```

```python
# let's create a file
with open('sample.txt','w') as f:
  f.write('hello world')
```

```python
# try catch demo
try:
  with open('sample2.txt','r') as f:
      print(f.read())
except:
   print("Sorry file not found!")
```

```python
# catching specific exception
try:
    m=5
    with open('sample1.txt','r') as f:
        print(f.read())
        print(m)
        #print(5/0)
        l=[0,1,2,3]
        l[100]
except FileNotFoundError:
    print("File not found!")
except NameError:
    print("variable not defined!")
except ZeroDivisionError:
    print("error division by zero is not allowed!")
except Exception as e: #if uhh did not provided the code or except block to some random error it gets handled by this
    print(e) # hamesha generic niche rhega and except upar nhi to generic will dominate
except Exception as e: 
    print(e.with_traceback) #tells name of the error 
```

```python
# else
try:
    f=open("sample.txt",'r')
except FileNotFoundError:
    print("File nhi mili")
except Exception:
    print("kuch to lafda hai")
else: #jo bhi kaam try ke sucess ke baad karna hai 
    #try ke code main kuch gadbad hua to exxcept main jayega 
    # and else main whi code aayega on which uhh are super sure code nhi fatega

    print(f.read())
```

```python
# finally
# kuch bhi ho finally main to aap jaoge hi cahhe error aaye ya naa aaye.. 
try:
    f=open("sample.txt",'r')
except FileNotFoundError:
    print("File nhi mili")
except Exception:
    print("kuch to lafda hai")
else: 
    print(f.read())
finally:
    print("ye to print hoga hi")
```

```python
#Exception are classes 
class Bank:
    def __init__(self,balance):
        self.balance=balance
    def withdraw(self,amount):
        if amount<0:
            raise Exception("Amount can not be less than zero!")
        elif self.balance<amount:
            raise Exception("paise nahi hain tere paas")
        else:
           self.balance=self.balance-amount
obj=Bank(10000)
try:
  obj.withdraw(15000) 
except Exception as e: #Exception ka object banaya hai 
    print(e)
else:
    print(obj.balance)
```

```python
class MyException(Exception): #inherit from Exception class karna padega for ur error class 
    def __init__(self,message):
        print(message)
class Bank:
    def __init__(self,balance):
        self.balance=balance
    def withdraw(self,amount):
        if amount<0:
            raise MyException("Amount can not be less than zero!")
        elif self.balance<amount:
            raise MyException("paise nahi hain tere paas")
        else:
           self.balance=self.balance-amount
obj=Bank(10000)
try:
  obj.withdraw(15000) 
except MyException as e: #Exception ka object banaya hai 
    pass
else:
    print(obj.balance)
```

```python
class SecurityError(Exception):
    def __init__(self,message):
        print(message)
    def logout(self):
        print("logout")
class Google:
    def __init__(self,name,email,password,device):
        self.name=self
        self.email=email
        self.password=password
        self.device=device
    def login(self,email,password,device):
        if device!=self.device:
            raise SecurityError('Bhai teri to lag gyi')
        if email==self.email and password==self.password:
            print("Welcome!")
        else:
            print("Login Error")
        

    
obj=Google('Mohit','ajay6378@gmail.com','1234','android')
try:
    obj.login('ajay6378@gmail.com','1234','windows')
except SecurityError as e: # logout security class ka object hai so call kar sakta hai
    e.logout()
else:
    print(obj.name)
finally:
    print("database connection closed!")
```
